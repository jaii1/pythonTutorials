```python
Before moving on, we recommend you to go through the HTML content of the webpage which we printed using soup.prettify() method and try to find a pattern or a way to navigate to the quotes. OR
NOW GO TO THE WEBSITE WHICH WE HAVE USED HERE AND RIGHT CLICK AND CHOSE INSPECT AND GO TO THE ELEMENTS, YOU WILL BE THERE
ONLY BY DEFAULT AND NOW LOOK FOR
It is noticed that all the quotes are inside a div container whose id is all_quotes'
we find that div element (termed as table) using find() method , see below
```

```python
import requests
import csv
from bs4 import BeautifulSoup
URL = "http://www.values.com/inspirational-quotes"
r = requests.get(URL)
soup = BeautifulSoup(r.content, 'html5lib')
print()

quotes = []
# to store quotes in list
soup.find('div', attr = {'id':'all_quotes'})

# now all quotes in div class are also again in a class(div class)i.e div class "col-6 col-lg-3.." so we iterate through this each 'div' to get
# quotes ,each iterating 'div' should have class quotes(div class), ok

for row in soup.find_all('div', attrs = {'class':'col-6 col-lg-3 text-center margin-30px-bottom sm-margin-30px-top'}):
#     print(row)
    quote = {}
    quote['theme'] = row.h5.text
    quote['url'] = row.a['href']
    quote['img'] = row.img['src']
    quote['lines'] = row.img['alt'].split(" #")[0]
    quote['author'] = row.img['alt'].split(" #")[1]
    quotes.append(quote)
    
filename = 'inspirational_quotes.csv'
with open(filename, 'w', newline='') as f:
    w = csv.DictWriter(f,['theme','url','img','lines','author'])
    w.writeheader()
    for quote in quotes:
        w.writerow(quote)


```

```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen
import html5lib

page = requests.get('https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off')
page

soup = BeautifulSoup(page.text, 'html.parser')
print(soup.prettify())
```

```python
#  div class="_3pLy-c row"

containers = soup.find_all('div', attrs = {'class':'_3pLy-c row'})

# Here this same 'class':'_3pLy-c row' will show detail of every phone which is on the page (in containers), ok so we have taken this class
# but we want here all the details of only first phone which is on the page, so we used 'container = containers[0]' and
# used prettify() with container

container  = containers[0]
print(container.prettify())

```

```python
# <div = class="_4rR01T"

# Now from the the first phone detail (which is on container),we want only phone name, so we used 'class:'_4rR01T'
# but this class will show every name of phone on that page so we used 'print(phone_name[0].text)', ok

phone_name =container.find_all('div', attrs = {'class':'_4rR01T'})
print(phone_name[0].text)

print()

# simililary price from first phone detail on page(which is in container)

price = container.find_all('div', attrs = {'class':'_30jeq3 _1_WHN1'})
print(price[0].text)

print()
# similarly rating from first phone detail on page(which is in container)

rating = container.find_all('div', attrs = {'class':'_3LWZlK'})
print(rating[0].text)
```

```python
all_phone_name = [item.find('div', attrs ={'class':'_4rR01T'}).get_text() for item in containers]
print(all_phone_name)

# similarly
print()

Phone_price = [item.find('div', attrs ={'class':'_30jeq3 _1_WHN1'}).get_text() for item in containers]
print(Phone_price)

print()

Phone_rating  =[item.find('div', attrs = {'class':'_3LWZlK'}).get_text() for item in containers]
print(Phone_rating)
                         
```

```python
import pandas as pd

detail_dict = {'Phone_name':all_phone_name, 'Price':Phone_price, 'Ratings':Phone_rating}
print()
phone_info = pd.DataFrame(detail_dict) 
phone_info
```

```python
phone_info.to_csv('phonedetail.csv')
```
