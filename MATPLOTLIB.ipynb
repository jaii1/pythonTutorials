```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
```

```python
so we will take 'x' values as numpy array of  a range up to 10 values and y as double of x
```

```python
x = np.arange(1,11)
x
```

```python
y = 2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
ADDING TITLE AND LABELS TO THE LINE PLOT
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.title('LINE PLOT')
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.show()
```

```python
ADDING COLOR,LINESTYLE,LINEWIDTH TO THE ABOVE EXISTING LINE PLOT
```

```python
plt.plot(x,y,color='orange',linestyle=':',linewidth=4)
plt.title('LINE_PLOT')
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.show()
```

```python
we can also use color red as 'r' and green as 'g'
```

```python
ADDING 2 LINES TO THE LINE PLOT
```

```python
x = np.arange(1,13)
x
```

```python
y1 =2*x
y1
```

```python
y2 = 3*x
y2
```

```python
plt.plot(x,y1,color='r',linestyle=":",linewidth=3)
plt.plot(x,y2,color='g',linestyle="-",linewidth=5)
plt.title('LINE_PLOT')
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.grid(True)
plt.show()
```

```python
BAR PLOT
```

```python
student = {"Bob":87,"Matt":56,"Sam":27,"jai":78}
names = list(student.keys())
marks =list(student.values())
plt.bar(names,marks)
plt.show() 
```

```python
ADDING TITLE & LABELS TO THE BAR PLOT
```

```python
plt.bar(names,marks)
plt.title("Bar_plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.show()
```

```python
HORIZONTAL BAR PLOT
```

```python
plt.barh(names,marks,color='r')
plt.title('Hrizontal_Bar_Plot')
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()     
```

```python
x = [10,20,30,40,50,60,70,80,90]
a =[8,1,7,2,0,3,7,3,2]
plt.scatter(x,a,marker="*",c='r',s=100)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,80,90]
a =[8,1,7,2,0,3,7,3,2]
b=[7,2,5,6,9,1,4,5,3]
plt.scatter(x,a,c="g",s=100)
plt.scatter(x,b,c="r",marker="*",s=200)
plt.title("SCATTER_PLOT")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
a =[8,1,7,2,0,3,7,3,2]
b=[7,2,5,8,1,4,9,4,2]
plt.subplot(1,2,1)
plt.scatter(x,a,color="g",s=200)
plt.title("Scatter_plot1")
plt.subplot(1,2,2)
plt.scatter(x,b,color="yellow",marker="*",s=200)
plt.title("Scatter_plot2")
plt.show()
```

```python
PIE-CHART
```

```python
fruits = ['Apple','Orange','Mango','Guava']
Quantity= [67,34,100,29]
plt.pie(Quantity,labels=fruits)
plt.show()
```

```python
fruits = ['Apple','Orange','Mango','Guava']
Quantity= [67,34,100,29]
plt.pie(Quantity,labels=fruits,autopct="%0.1f%%",colors=["green","yellow","red","pink"])
plt.show()
```

```python
DOUGHNUT-CHART
```

```python
plt.pie(Quantity,labels=fruits,radius=2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
