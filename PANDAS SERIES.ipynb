```python
# CREATING DATAFRAME FROM ZIP FUNCTION(listoftuple)
Name = ['tom', 'krish', 'nick', 'juli'] 
Age = [25, 30, 26, 22]
listoftuple = list(zip(Name, Age))
# IN ZIP FUNCTION,INNER VALUES ALWAYS RETURN IN TUPLE
print(listoftuple)
df = pd.DataFrame(listoftuple)
df
df2 = pd.DataFrame(listoftuple, columns =["Name", "Age"])
df2
```

```python
#  CREATING DATAFRAME FROM ZIP FUNCTION(tupleoftuple)
Name = ['tom', 'krish', 'nick', 'juli'] 
Age = [25, 30, 26, 22]
tupleoftuple = tuple(zip(Name, Age))
print(tupleoftuple)
df = pd.DataFrame(tupleoftuple, columns = ["Name","Age"])
df
```

```python
#  CREATING DATAFRAME FROM ZIP FUNCTION(setoftuple)
Name = ["jai","sam","ram"]
Age = [34,35,37]
setoftuple = set(zip(Name,Age))
setoftuple
df = pd.DataFrame(setoftuple, columns = ["Name","Age"])
df
```
