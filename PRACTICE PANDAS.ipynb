```python
x =pd.Series({"a":23, "b":(9,0),100:[5,6],"n":"jai"})
print("x:", x)
print(type(x))
print()
x1 = pd.Series({"a":20, "b":30, "c":40})
print("x1:", x1)
print()
x2 = pd.Series({"m":10,"n":20,"k":40}, index=["k","m","p","n"])
print("x2:", x2)
print()
x4 = pd.Series({"emp":"jai","salary":45000,"date":"24oct","age":34,(9,0):[7,8]}, index = ["salary",(9,0),"gender","age","date"])
print("x4:", x4)
```

```python
EXTRACTING VALUES FROM  GIVEN SERIES
```

```python
s1 = pd.Series([1,2,3,4,5,"j",(7,8),5.6,11])
print("s1:", s1)
print()
s2 = s1[:6]
print("s2", s2)
print()
s3 = s1[-3:]
print("s3:", s3)
print()
s4 = s1[4:]
print("s4:", s4)
print()
s5 = s1[:-4]
print("s5:", s5)
print()
s6 = s1[0:7:3]
print("s6:", s6)
print()
s7 = s1[1::4]
print("s7:", s7)
```
