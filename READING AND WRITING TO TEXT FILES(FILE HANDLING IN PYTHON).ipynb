```python
Text file mode
1)"r" = used for open for reading.if not found then it shows 'FileNotFoundError'
2)"w" = It open the file to write , if any data already there in file then it overwrite it and if file doesnt exists 
then it create the file
3)"x":=it will also create file  and write data  but file should not be availiabe,if it already exists then it will show
FileExistsError
4)a:= It will append data at the end of the file if it exists, otherwise it will create and write
5)r+ :=open for reading and then writing
6) w+:=open for writing and then reading,it will overwrite existing data
7)a+:=open for appending and then reading,it won't overwrite the data
```

```python
f = open('student2.text', mode='r+')
data = f.read()
f.write(" machine learning")
print(data)
f.close()

In this 'r+' it has first read the file and print here and then it write to the same file('machine learnig')which you
see by opening the file, but cannot see here, bcoz first it read the content of file and then write
```

```python
w+ =It will first writing to the file and then reading to file ,if file already there then it will overwrite it
here we will take 'student4.text'
```

```python
f = open('student4.text', mode='w+')
print(f.tell())
f.write("python is a great language")
print(f.tell())
data = f.read()
print(f.tell())
print(data)
print(f.tell())
f.close()
print("end")
# now here position is at 26 after writing something into file as you can see ,so when you try to read something then it will
# show you nothing, so we use seek() method to move to first position so that it can read the file
```

```python
f = open('student4.text', mode='w+')
print(f.tell())
f.write("I want to learn Machine Learning")
print(f.tell())
f.seek(0)
print(f.tell())
data = f.read()
print(f.tell())
print(data)
print(f.tell())
f.close()
print("finish")
```

```python
'a+' =it will append data at the ennd and then read the data..it will not overwrite,but if file is not there then
it will create and write data in it
```

```python
f = open('student4.text', mode='a+')
print(f.tell())
f.write(' pandas used for data analysis in python')
f.seek(0)
print(f.tell())
data = f.read()
print(f.tell())
print(data)
print(f.tell())
f.close()
print("end")
```
